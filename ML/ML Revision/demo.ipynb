```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('student_Scores.csv')
print(df.head)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
plt.scatter(df['Hours'],df['Scores'],c='blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores Obtained')
plt.show()
```

```python
x=df[['Hours']]
y=df['Scores']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
plt.scatter(x,y,c='blue')
plt.plot(x,model.predict(x),color='red')
plt.title('Regression Line: Hours vs Scores')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```

```python
# Function to predict score based on user input hours
def predict_score(hours):
    score = model.predict([[hours]])[0]
    return round(score, 2)

# Example predictions
example_hours = [1.5, 5, 10]
predictions = {h: predict_score(h) for h in example_hours}
predictions
```
